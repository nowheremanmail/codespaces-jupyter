### Introducción
El conjunto de datos con el que vamos a trabajar (`ThunderbeamData.json`)
consta de las páginas registradas con Lightbeam, la extensión del navegador
que utiliza visualizaciones interactivas para mostrarnos las relaciones entre 
terceros y las páginas que visitamos. 

```python
import pandas as pd
import json

# Generamos un dataframe con los datos que hemos descargado del Lightbeam.
df = pd.read_json('../data/ThunderbeamData.json')
df = df.transpose()

# # Restringimos los registros a las páginas que hemos visitado.
pagina = df[df.firstParty == True]
```

```python
# Imprimimos las columnas del dataframe que hemos cargado.
print(pagina.columns.values)

# Imprimimos la lista de páginas que hemos visitado.
print(pagina.hostname)
```

```python
# Generamos las parejas de páginas, la lista de terceros en común, y cuántos son.
PaginesTercers = [[pagina.hostname[u], 
                   pagina.hostname[v], 
                   list(set(pagina.thirdParties[u]) & set(pagina.thirdParties[v])), 
                   len(list(set(pagina.thirdParties[u]) & set(pagina.thirdParties[v])))]
                  for u in pagina.index 
                  for v in pagina.index
                  if u != v]

# Convertimos los datos que hemos calculado en un dataframe.
PaginesTercers = pd.DataFrame(PaginesTercers, 
                              columns = ['Pagina1', 
                                         'Pagina2', 
                                         'TercersEnComu', 
                                         'Interseccio'])  
```

```python
# Calculamos cuáles son las páginas que tienen más terceros en común.
PaginesTercers.iloc[PaginesTercers.Interseccio.idxmax()]
```

```python
ordenados_por_inter = PaginesTercers.sort_values(['Interseccio'], ascending=[False])
print(ordenados_por_inter[['Pagina1','Pagina2','Interseccio']])

# mostramos los 3 primeros
print(ordenados_por_inter.loc[13])
print(ordenados_por_inter.loc[336])
print(ordenados_por_inter.loc[339])
```

```python
from functools import reduce

terceros = reduce(lambda a,b: set(a) | set(b), [set(pagina.thirdParties[v]) for v in pagina.index])

print(terceros)

print("tamaños de el conjunto de terceros únicos " + str(len(terceros)))
```

```python
# calculamos columna con el número de third parties
df['numero_third_parties'] = df.apply(lambda row: len(row['thirdParties']), axis=1)

# ordenamos descendientemente por el número, filtrando con páginas visitadas
ordenado = df.sort_values(['numero_third_parties'], ascending=[False])[df.firstParty == True]

print (ordenado[['hostname','numero_third_parties']])
```

```python
from functools import reduce
import statistics
from statistics import mode

lista_todos_terceros = reduce(lambda a,b: a + b, [pagina.thirdParties[v] for v in pagina.index])

print(lista_todos_terceros)

print("tamaño lista de todos los terceros " + str(len(lista_todos_terceros)))

print("el tercero mas frecuente es " + mode(lista_todos_terceros))


```
